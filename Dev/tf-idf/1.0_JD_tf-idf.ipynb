```python


def format_documents(doc):
    doc = doc.lower()
    doc = doc.translate(str.maketrans('', '', string.punctuation))
    doc = doc.split(' ')
    
    return doc

def inverse_document_frequency(docs,tokens,num_docs):
    D = defaultdict(int)
    for token in tokens:
        for doc in docs:
            if token in doc:
                D[token] += 1
    
    idf_dict = defaultdict(float)
    for tok,count in D.items():
        idf_dict[tok] = log(num_docs/count)
    return idf_dict

def tfidfVectorizer(query):
    f_query = format_documents(query)
    query_tf = Counter(f_query)
    query_dict  = defaultdict(float)
    for key in query_tf.keys():
        query_dict[key] += idf[key]*query_tf[key]
    return query_dict
    
```
